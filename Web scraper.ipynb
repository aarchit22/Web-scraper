What is shell?
Shell is a program that provides a command-line interface for interacting with an operating system, while bash, a term often used interchangeably with shell, is actually one of the most commonly used Linux shells.

Bash scripting:
Bash scripting, on the other hand is a very powerful tool for automating system tasks, managing system resources and performing other repetitive tasks in Linux systems. This is done through a bash script which is a file containing a sequence of commands that are executed by the bash program line by line. These sequence of commands can be run multiple times by executing the script.

Some basic bash commands are:-

pwd: gives the present working directory(folder, terms of GUI)

ls: lists all the files and sub-directories present in present working directory

ls -a: lists all the files, even the hidden ones

ls -l: gives detailed information about the files and directories in current directory

cd: to enter a directory

cd .. :goes to the parent directory of current directory

touch: to create a new file

vim, nano: text editors to open a  file for editing

man, help: generally used as a manual for other commands

chmod: to update the access/ permissions of a file given to the owner, group and other users

echo: to display something on terminal, also sometimes used to pass values to a file or variable

cat: to add contents to a file or even from one file to another

mkdir/rmdir: to create/remove a directory

rm: to remove a file or a directory

cp: copy contents from one file/directory to another

sudo: a very important command which grants total access to our system as a root user

Now I will be building a link scraper for any webpage using Python and storing its output in a file for future use.

```python
# importing all the necessary libraries

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import codecs
import re
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# sets up and initializes a Chrome driver
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
# website to be scraped
val="https://moviesmod.biz/"
# creating a WebDriverWait instance
wait = WebDriverWait(driver, 10)
```

```python
# Opens the epecified url in val in a new browser window
driver.get(val)
```

```python
# it gets the current url and checks if it is the one we were meant to be at, and retrieves the HTML source code of the currently loaded web page
get_url = driver.current_url
wait.until(EC.url_to_be(val))
if get_url == val:
    page_source = driver.page_source
```

```python
# parses the HTML source code stored in page_source using the html.parser
soup = BeautifulSoup(page_source,features="html.parser")
```

```python
# finds all anchor tags in the html body
anchors=soup.body.find_all("a")
```

```python
# opens a new file in append mode and stores all the href values of anchor tags
file=codecs.open('linkss.txt','a+')
file.write("Following are all the links:\n")
for i in anchors:
    link=i.get("href")
    file.write(str(link)+"\n")

file.close()
```

```python
# closes all the browser windows
driver.quit()
```

```python
# Printing the content of linkss.txt file
with open('linkss.txt', 'r') as file:
    content = file.read()
print(content)
```

This will be a word scraper to scrape off all the instances and the related context of keyword in an article on a website

```python
# importing all the necessary libraries

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import codecs
import re
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# sets up and initializes a Chrome driver
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
wait = WebDriverWait(driver, 10)
```

```python
# website to be scraped
val="https://www.browserstack.com/guide/how-ai-in-visual-testing-is-evolving"
```

```python
# Opens the epecified url in val in a new browser window
driver.get(val)
```

```python
# it gets the current url and checks if it is the one we were meant to be at, and retrieves the HTML source code of the currently loaded web page
get_url = driver.current_url
wait.until(EC.url_to_be(val))
if get_url == val:
    page_source = driver.page_source
```

```python
# parses the HTML source code stored in page_source using the html.parser
soup = BeautifulSoup(page_source,features="html.parser")
```

```python
# the keyword whose intances are to be found
keyword=input("Enter a keyword to find instances of in the article:")
```

```python
# finds all the occurrences of the entered keyword
matches = soup.body.find_all(string=re.compile(keyword))
```

```python
# adds the relevant information regarding occurrences of keyword in the webpage to the file
len_match = len(matches)
title = soup.title.text
file=codecs.open('article_scraping.txt', 'a+')
file.write(title+"\n")
file.write("The following are all instances of your keyword:\n")
count=1
for i in matches:
    file.write(str(count) + ". " +  i  + "\n")
    count+=1
```

```python
file.write("There were "+str(len_match)+" matches for the keyword.")
```

```python
file.close()
driver.quit()
```

This is an image scraper to  scrape images from a webpage

```python
# import all the necessary python libraries

from selenium import webdriver
from bs4 import BeautifulSoup
from PIL import Image
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import os
import matplotlib.pyplot as plt
```

```python
url = 'https://moviesmod.biz/'  # the URL of the webpage
folder_name = 'images'  # Name of the folder to save images
```

```python
# Initialize Selenium webdriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
wait = WebDriverWait(driver, 10)
```

```python
# Load the webpage
driver.get(url)
```

Note: Don't close the newly opened browser tab

```python
# Get the page source after loading
get_url = driver.current_url
wait.until(EC.url_to_be(url))
if get_url == url:
    page_source = driver.page_source
```

```python
# Parse the HTML content of the webpage
soup = BeautifulSoup(page_source, 'html.parser')
```

```python
# Create a folder to save the images if it doesn't exist
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
```

```python
# Find all img tags in the HTML content
tags = soup.find_all('img')
```

```python
# Download and save each image to the folder
for index, tag in enumerate(tags, 1):
    img_url = tag['src']
    img_name = f"image_{index}.png"
    img_path = os.path.join(folder_name, img_name)

    # Download the image using Selenium
    driver.get(img_url)

    # Capture the screenshot and save it as an image
    driver.save_screenshot(img_path)
    print(f"Downloaded: {img_name}")

    # Display the image
    img = Image.open(img_path)
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
# Close the webdriver
driver.quit()
```

Below is the bash scipt to extract the names, date of births and pictures of Prime Ministers of India, but it must be executed in a shell environment for proper functioning.

```python
# jq is  a json processor to manipulate json data directly from the command-line
!apt-get install jq
```

```python
# To upload the xidel  binary file to /content/ directory of colab notebook
from google.colab import files
uploaded=files.upload()
```

```python
%%bash
#!/bin/bash

# Sets the Internal Field Separator to newline
IFS=$'\n'

# Defines the URL to scrape
url='https://en.wikipedia.org/wiki/List_of_prime_ministers_of_India'

# Defines the XPath expressions of sublinks to be explored from that webpage
crawl_elems='//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[3]/b/a'

# XPath expressions of data to be extracted from the webpage
xpath_name='//span[@class="mw-page-title-main"]'
xpath_bday='//span[@class="bday"]'
xpath_image='//span[@class="mw-default-size"]/a/img/@src'

# Path to the Xidel binary executable in colab notebook
xidel_path="/content/xidel"

# Executes Xidel to scrape the webpage and output JSON data
# uses jq to extract specific elements from the JSON
# '-e' is used to specify XPath expressions for data extraction
# '-f' is used to specify/filter the XPath expressions relevant to those mentioned in 'crawl_elems'
data=($("$xidel_path" "$url" -f "$crawl_elems" -e "$xpath_name" -e "$xpath_bday" -e "$xpath_image" --output-format=json-wrapped | jq -r .[]))

# Loops through the extracted data and write it to a CSV file
for ((i=0; i < ${#data[@]}; i+=3));do

        # Checks if the birthdate is null and skip if it is
        if [ "${data[i+1]}" = "null" ]; then continue; fi

        # Writes the name, birthdate, and image URL to the CSV file
        echo ${data[i]},${data[i+1]},${data[i+2]} >> data.csv
done
```
